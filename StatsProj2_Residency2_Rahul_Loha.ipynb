1. Find the mean of old scheme and new scheme column.

```python
old_scheme_mean = payment_scheme['Old Scheme'].mean()
new_scheme_mean = payment_scheme['New Scheme'].mean()
print("Mean of Old scheme is : ", old_scheme_mean)
print("Mean of New scheme is : ", new_scheme_mean)
```

2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?

Null Hypothesis : Old scheme mean is equal to New scheme mean
    
Alternative Hypothesis : Mean of new scheme is greater than mean of old scheme

Sample size is 30 so I am using  t-test here.
For Hypotesis analysis I am using two sampled paired t-test because here we are checking the effect of new scheme over the existing scheme

```python
from scipy import stats
oldscheme = payment_scheme['Old Scheme']
newscheme = payment_scheme['New Scheme']
# print(oldscheme, newscheme)
payment_scheme[['Old Scheme','New Scheme']].describe()
ttset, pval = stats.ttest_rel(payment_scheme['Old Scheme'], payment_scheme['New Scheme'])
```

```python
print("P-Value",pval)
```

```python
if pval < 0.05:
    print("Rejecting Null Hypothesis")

if pval > 0.05:
    print("Accepting Null Hypothesis")
```

Output: New scheme has not raised the sales. New scheme to increase the sales has failed

3. What conclusion does the test (p-value) lead to?

P value is higher than the 0.05 ( 95% confidence level) that shows that null hypothesis is true. 
So we can conclude that new scheme is not significantly increasing the sales. 
So we should not consider this new scheme.
