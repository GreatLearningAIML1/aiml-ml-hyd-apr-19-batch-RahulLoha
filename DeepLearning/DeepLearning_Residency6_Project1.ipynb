2. Drop the columns which are unique for all users like IDs

```python
#Removing RowNumber, CustomerId and Surname because they are unique for every customer and will not effect with prediction.
df_bank = df_bank.drop(['RowNumber','CustomerId','Surname'], axis =1)
```

```python
df_bank.head()
```

3. Distinguish the feature and target set

```python
features = df_bank.drop(['Exited'], axis =1)
target = df_bank['Exited']
```

```python
features.info()
```

Label Encoding for variables with object type

```python
from sklearn import preprocessing

le1 = preprocessing.LabelEncoder()
le2 = preprocessing.LabelEncoder()

features["Geography"] = le1.fit_transform(features["Geography"])
features["Gender"] = le2.fit_transform(features["Gender"])
```

```python
features.info()
```

```python
#Spilitting the Dataset into training and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3)
```

```python
#Normalizing the features

from sklearn.preprocessing import StandardScaler
z = StandardScaler()
X_train = z.fit_transform(X_train)
X_test = z.transform(X_test)
```

```python
X_train.shape
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initializing Neural Network
classifier = Sequential()
# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 10))
# Adding the second hidden layer
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling Neural Network
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting our model 
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Creating the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_pred)
```
